<h4>What is Discrete variable?</h4>

<h3>WHAT ARE INDEPENDENT AND DEPENDENT VARIABLES?</h3>

<h4>What's an independent variable?</h4>

>An independent variable is exactly what it sounds like. It is a variable that stands alone and isn't changed by the other variables you are trying to measure. For example, someone's age might be an independent variable. Other factors (such as what they eat, how much they go to school, how much television they watch) aren't going to change a person's age.

<h4>What's an dependent variable?</h4>

>Just like an independent variable, a dependent variable is exactly what it sounds like. It is something that depends on other factors. For example, a test score could be a dependent variable because it could change depending on several factors such as how much you studied, how much sleep you got the night before you took the test, or even how hungry you were when you took it.

<h3>Difference between dependent and independent variable</h3>

>The main key difference between these two<br> A dependent variable is a variable whose variations depend on another variableusually the independent variable.<br> on the other hand <br> An Independent variable is a variable whose variations do not depend on another variable.
